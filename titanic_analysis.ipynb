```python
import pandas as pd

# Create dictionary
data = {
    "Name": ["Abel", "Sara", "John", "Muna", "David", "Liya", "Noah", "Ruth", "Eden", "Samuel",
             "Hana", "Ben", "Marta", "Kaleb", "Selam"],
    "Age": [22, 24, 21, 23, 25, 22, 20, 24, 23, 26, 22, 21, 24, 23, 25],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT", "Finance", "HR", "IT", "Finance",
                   "HR", "IT", "Finance", "HR", "IT"],
    "Salary": [5000, 4500, 5200, 4800, 4600, 5100, 4700, 4550, 5300, 4900,
               4400, 5250, 4750, 4600, 5400],
    "Experience": [2, 3, 1, 2, 4, 2, 1, 3, 2, 5, 2, 1, 3, 4, 5]
}

# Custom index
index_labels = [f"EMP{i}" for i in range(1, 16)]

df = pd.DataFrame(data, index=index_labels)

print(df)
```

```python
titanic = pd.read_csv("/content/Titanic-Dataset.csv")
```

 Exploration

```python
print(titanic.head())
print(titanic.info())
print(titanic.describe())
```

**Data Cleaning**

Handle missing values:

```python
# Fill Age with median
titanic["Age"].fillna(titanic["Age"].median(), inplace=True)

# Fill Embarked with mode
titanic["Embarked"].fillna(titanic["Embarked"].mode()[0], inplace=True)

# Drop Cabin column
titanic.drop(columns=["Cabin"], inplace=True)

# Remove duplicates
titanic.drop_duplicates(inplace=True)
```

**Data Analysis (groupby)**

Survival rate by gender

```python
survival_gender = titanic.groupby("Sex")["Survived"].mean()
print(survival_gender)
```

Survival rate by class

```python
survival_class = titanic.groupby("Pclass")["Survived"].mean()
print(survival_class)
```

Average age per class

```python
avg_age_class = titanic.groupby("Pclass")["Age"].mean()
print(avg_age_class)
```

Survival rate by age group

```python
# Create age groups
bins = [0, 12, 18, 35, 60, 100]
labels = ["Child", "Teen", "Young Adult", "Adult", "Senior"]

titanic["AgeGroup"] = pd.cut(titanic["Age"], bins=bins, labels=labels)

survival_age_group = titanic.groupby("AgeGroup")["Survived"].mean()
print(survival_age_group)
```

**Filtering**

Female passengers who survived

```python
female_survived = titanic[(titanic["Sex"] == "female") & (titanic["Survived"] == 1)]
print(female_survived)
```

Children who survived

```python
children_survived = titanic[(titanic["Age"] < 12) & (titanic["Survived"] == 1)]
print(children_survived)
```

First class passengers with high survival probability

```python
first_class_survived = titanic[(titanic["Pclass"] == 1) & (titanic["Survived"] == 1)]
print(first_class_survived)
```

# **Insights**


**1. Who was more likely to survive?**

Females were much more likely to survive than males.


* Survival rate of females is significantly higher.
* This reflects the “women first” rescue policy.

 Yes, passenger class strongly affected survival.

* 1st class → highest survival rate

**3. Were children prioritized?**

Yes, children had higher survival rates.


* Many children fall into the “Child” age group.

Female + 1st Class + Young/Child
